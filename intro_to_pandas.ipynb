```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
avg_monthly_precip = pd.DataFrame(columns=["month", "precip_in"],
                                  data=[
                                      ["Jan", 0.70],  ["Feb", 0.75],
                                       ["Mar", 1.85],  ["Apr", 2.93],
                                       ["May", 3.05],  ["June", 2.02],
                                       ["July", 1.93], ["Aug", 1.62],
                                       ["Sept", 1.84], ["Oct", 1.31],
                                       ["Nov", 1.39],  ["Dec", 0.84]
                                  ])

avg_monthly_precip
```

```python
f, ax = plt.subplots()
avg_monthly_precip.plot(x="month",
                        y="precip_in",
                        title="Plot of pandas DataFrame using pandas .plot",
                        ax=ax)
plt.show()
```

```python
f, ax = plt.subplots()

ax.plot(avg_monthly_precip.month,
        avg_monthly_precip.precip_in)

ax.set(title="Another way to plot stuff with pandas")
plt.show()
```
